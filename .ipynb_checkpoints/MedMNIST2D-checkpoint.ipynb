# 2. Algorithm design and setup, Data Processing & Image Augmentation

We first built basic baseline models, including MLP and CNN, to establish reference performance before any enhancements. These models allowed us to evaluate the effectiveness of different image augmentation techniques in a controlled setup. Although Gradient Boosting (XGBoost) was initially included as part of the baseline comparison, we excluded it from the augmentation experiments due to its significantly longer training time.

## 2.1 build baseline model

We use the baseline model as a reference point to evaluate the effectiveness of each augmentation technique. 
The primary evaluation metric is classification accuracy on the test set.
the result of using the training data directly without any data processing is: 

+ MLP Accuracy: 14.68%
+ CNN Accuracy: 69.76%
+ Gradient Boost: 42.06%

```python
def model_training(X_train, X_test, y_train, y_test): 
    # Train/Val split for deep learning models
    X_train_, X_val, y_train_, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

    num_classes = len(np.unique(y_train))
    y_train_cat = to_categorical(y_train, num_classes)
    y_test_cat = to_categorical(y_test, num_classes)
    y_train_cat_ = to_categorical(y_train_, num_classes)
    y_val_cat = to_categorical(y_val, num_classes)

    # ------------------ MLP ------------------
    mlp_model = Sequential([
        Flatten(input_shape=X_train.shape[1:]),
        Dense(256, activation='relu'),
        Dropout(0.3),
        Dense(128, activation='relu'),
        Dropout(0.3),
        Dense(num_classes, activation='softmax')
    ])
    
    mlp_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    mlp_model.fit(X_train_, y_train_cat_, validation_data=(X_val, y_val_cat), epochs=10, batch_size=32, verbose=0)
    
    mlp_eval = mlp_model.evaluate(X_test, y_test_cat, verbose=0)
    print(f"MLP Accuracy: {mlp_eval[1]:.4f}")

    # ------------------ CNN ------------------
    cnn_model = Sequential([
        Conv2D(32, (3, 3), activation='relu', input_shape=X_train.shape[1:]),
        MaxPooling2D((2, 2)),
        Conv2D(64, (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),
        Conv2D(128, (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),
        Flatten(),
        Dense(128, activation='relu'),
        Dropout(0.3),
        Dense(num_classes, activation='softmax')
    ])
    
    cnn_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    cnn_model.fit(X_train_, y_train_cat_, validation_data=(X_val, y_val_cat), epochs=10, batch_size=32, verbose=0)

    cnn_eval = cnn_model.evaluate(X_test, y_test_cat, verbose=0)
    print(f"CNN Accuracy: {cnn_eval[1]:.4f}")

    # ------------------ XGBoost ------------------
    # Flatten image data for XGBoost
    X_train_flat = X_train.reshape(X_train.shape[0], -1)
    X_test_flat = X_test.reshape(X_test.shape[0], -1)

    xgb_model = XGBClassifier(n_estimators=2, max_depth=2, learning_rate=1, use_label_encoder=False, eval_metric='mlogloss')
    xgb_model.fit(X_train_flat, y_train)

    xgb_preds = xgb_model.predict(X_test_flat)
    xgb_acc = accuracy_score(y_test, xgb_preds)
    print(f"XGBoost Accuracy: {xgb_acc:.4f}")
```

```python
model_training(X_train, X_test, y_train, y_test)
```

## 2.2  Data Processing + Image Augmentation 

### 2.2.1 Channel-wise Normalization
Normalize each RGB channel separately using z-score normalization

```python
def channel_wise_normalization(X_train, X_test):
    X_train_norm = X_train.astype('float32')/255.0
    X_test_norm = X_test.astype('float32')/255.0
    
    for channel in range(3):
        channel_mean = np.mean(X_train_norm[:, :, :, channel])
        channel_std = np.std(X_train_norm[:, :, :, channel])
        X_train_norm[:, :, :, channel] = (X_train_norm[:, :, :, channel] - channel_mean) / (channel_std + 1e-7)
        X_test_norm[:, :, :, channel] = (X_test_norm[:, :, :, channel] - channel_mean) / (channel_std + 1e-7)
    
    return X_train_norm, X_test_norm
```

```python
X_train_RGB_normal, X_test_RGB_normal = channel_wise_normalization(X_train, X_test)
model_training(X_train_RGB_normal, X_test_RGB_normal, y_train_cat, y_test_cat)
```

### 2.2.2 Color Deconvolution
Color deconvolution is a technique used mainly in histopathology and microscopy image analysis to separate overlapping stains in an image into individual color channels.
simplify from: https://github.com/mchelali/Color-Deconvolution

```python
def color_deconvolution(X_train, X_test):
    X_train_dec = np.zeros_like(X_train, dtype=np.float32)
    X_test_dec = np.zeros_like(X_test, dtype=np.float32)
    
    def process_batch(batch):
        batch_float = batch.astype(np.float32) / 255.0
        result = np.zeros_like(batch_float)
        
        # Simple transformation based on color ratios
        # Channel 1: R dominance (R / (G+B+1))
        result[:, :, :, 0] = batch_float[:, :, :, 0] / (batch_float[:, :, :, 1] + batch_float[:, :, :, 2] + 0.01)
        
        # Channel 2: B dominance (B / (R+G+1))
        result[:, :, :, 1] = batch_float[:, :, :, 2] / (batch_float[:, :, :, 0] + batch_float[:, :, :, 1] + 0.01)
        
        # Channel 3: G intensity relative to mean
        result[:, :, :, 2] = batch_float[:, :, :, 1] / (np.mean(batch_float[:, :, :, :], axis=3) + 0.01)
        
        # Clip to handle potential outliers
        return np.clip(result, 0, 1)
    
    X_train_dec = process_batch(X_train)
    X_test_dec = process_batch(X_test)
    
    return X_train_dec, X_test_dec
```

```python
X_train_dec, X_test_dec = color_deconvolution(X_train, X_test)
X_train_normal, X_test_normal = channel_wise_normalization(X_train_dec, X_test_dec)
model_training(X_train_normal, X_test_normal, y_train_cat, y_test_cat)
```

### 2.2.3 CLAHE
Contrast-Limited Adaptive Histogram Equalization is an image processing technique used to enhance contrast in images, particularly in areas with low contrast or high noise 
https://docs.opencv.org/4.x/d5/daf/tutorial_py_histogram_equalization.html

```python
def clahe(X_train, X_test):
    X_train_eq = np.zeros_like(X_train, dtype=np.float32)
    X_test_eq = np.zeros_like(X_test, dtype=np.float32)
    
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(4, 4))
    
    for i in range(X_train.shape[0]):
        for c in range(3):
            X_train_eq[i, :, :, c] = clahe.apply(X_train[i, :, :, c].astype(np.uint8))/ 255.0
    
    for i in range(X_test.shape[0]):
        for c in range(3):
            X_test_eq[i, :, :, c] = clahe.apply(X_test[i, :, :, c].astype(np.uint8))/255.0
    
    return X_train_eq, X_test_eq
```

```python
X_train_clahe, X_test_clahe = clahe(X_train, X_test)
X_train_normal, X_test_normal = channel_wise_normalization(X_train_clahe, X_test_clahe)
model_training(X_train_normal, X_test_normal, y_train_cat, y_test_cat)
```

### 2.2.4 HSV Color Space
HSV color space contains color (Hue, H), Saturation (S), and light (Value, V), which lets the data reflect more details. with cv2 library.
https://docs.opencv.org/4.x/df/d9d/tutorial_py_colorspaces.html

```python
def convert_to_hsv(X_train, X_test):
    X_train_hsv = np.zeros_like(X_train, dtype=np.float32)
    X_test_hsv = np.zeros_like(X_test, dtype=np.float32)
    
    for i in range(X_train.shape[0]):
        X_train_hsv[i] = cv2.cvtColor(X_train[i].astype(np.uint8), cv2.COLOR_RGB2HSV) / 255.0
    
    for i in range(X_test.shape[0]):
        X_test_hsv[i] = cv2.cvtColor(X_test[i].astype(np.uint8), cv2.COLOR_RGB2HSV) / 255.0
    
    return X_train_hsv, X_test_hsv
```

```python
X_train_hsv, X_test_hsv = convert_to_hsv(X_train, X_test)
X_train_normal, X_test_normal = channel_wise_normalization(X_train_hsv, X_test_hsv)
model_training(X_train_normal, X_test_normal, y_train_cat, y_test_cat)
```

### 2.2.5 LAB Color Space
LAB represents Lightness (L) and chromatic components (A and B). The separation can be more similar to human perception and is more useful for classification tasks where human-like interpretation of color is needed. 
with cv2 library. 
https://docs.opencv.org/3.4/de/d25/imgproc_color_conversions.html

```python
def convert_to_lab(X_train, X_test):
    X_train_lab = np.zeros_like(X_train, dtype=np.float32)
    X_test_lab = np.zeros_like(X_test, dtype=np.float32)
    
    for i in range(X_train.shape[0]):
        X_train_lab[i] = cv2.cvtColor(X_train[i].astype(np.uint8), cv2.COLOR_RGB2Lab) / 255.0
       
    for i in range(X_test.shape[0]):
        X_test_lab[i] = cv2.cvtColor(X_test[i].astype(np.uint8), cv2.COLOR_RGB2Lab) / 255.0
    
    return X_train_lab, X_test_lab
```

```python
X_train_lab, X_test_lab = convert_to_lab(X_train, X_test)
X_train_normal, X_test_normal = channel_wise_normalization(X_train_lab, X_test_lab)
model_training(X_train_normal, X_test_normal, y_train_cat, y_test_cat)
```

### 2.2.6 Sobel edge
Add Sobel edge feature as the 4th channel to normalized RGB images. 
our code is generated with the help of: 
+ cv2 document: https://docs.opencv.org/4.x/d5/d0f/tutorial_py_gradients.html
+ normalization: https://gist.github.com/hackintoshrao/cc0e10f58079ede6866111d6ed75b420

```python
def extract_edge_features(X_train_rgb_norm, X_test_rgb_norm):
    """
    Parameters:
    - X_train_rgb_norm: normalized training images, shape (N, H, W, 3)
    - X_test_rgb_norm: normalized test images, shape (N, H, W, 3)
    
    Returns:
    - X_train_combined: shape (N, H, W, 4)
    - X_test_combined: shape (N, H, W, 4)
    """
    def compute_edge(img):
        gray = cv2.cvtColor((img * 255).astype(np.uint8), cv2.COLOR_RGB2GRAY)
        sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
        sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)
        edge = np.sqrt(sobelx**2 + sobely**2)
        edge = edge / (edge.max() + 1e-7)
        return edge.astype(np.float32)

    N_train, H, W, _ = X_train_rgb_norm.shape
    N_test = X_test_rgb_norm.shape[0]

    X_train_combined = np.zeros((N_train, H, W, 4), dtype=np.float32)
    X_test_combined = np.zeros((N_test, H, W, 4), dtype=np.float32)

    X_train_combined[:, :, :, :3] = X_train_rgb_norm
    X_test_combined[:, :, :, :3] = X_test_rgb_norm

    # Add edge channel
    for i in range(N_train):
        X_train_combined[i, :, :, 3] = compute_edge(X_train_rgb_norm[i])
    for i in range(N_test):
        X_test_combined[i, :, :, 3] = compute_edge(X_test_rgb_norm[i])

    return X_train_combined, X_test_combined
```

```python
X_train_edge, X_test_edge = extract_edge_features(X_train, X_test)
X_train_normal, X_test_normal = channel_wise_normalization(X_train_edge, X_test_edge)
model_training(X_train_normal, X_test_normal, y_train_cat, y_test_cat)
```

## Fully connected neural network

```python
# Early stopping callback to prevent overfitting during tuning
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=3,
    restore_best_weights=True
)

# --------------------------------
# MLP Hyperparameter Tuning
# Number of dense layers (3-5)
# Units per layer (64-512)
# Dropout rates (0.3-0.4)
# Learning rates (0.001, 0.0001)
# --------------------------------

def build_mlp_model(hp):
    model = Sequential()
    model.add(Flatten(input_shape=X_train.shape[1:]))

    num_layers = hp.Int('num_layers', 3, 4)

    for i in range(num_layers):
        units = hp.Choice(f'units_{i}', [128, 256, 512])
        dropout_rate = hp.Choice(f'dropout_{i}', [0.3, 0.4])
        model.add(Dense(units, activation='relu'))
        model.add(BatchNormalization())
        model.add(Dropout(dropout_rate))

    model.add(Dense(num_classes, activation='softmax'))

    learning_rate = hp.Choice('learning_rate', [1e-3, 1e-4])
    optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)

    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
    return model


# --------------------------------
# Run the tuners
# --------------------------------

def run_hyperparameter_tuning(mlp_tuner, X_train_, y_train_, X_val, y_val):
    # For debugging: print input shapes
    print(f"X_train shape: {X_train_.shape}")
    print(f"y_train shape: {y_train_.shape}")

    try:
        print("Starting MLP hyperparameter tuning...")
        mlp_start = time.time()
        mlp_tuner.search(
            X_train_, y_train_,
            validation_data=(X_val, y_val),
            epochs=10,
            batch_size= 32,
            callbacks=[early_stopping],
            verbose=1
        )

        best_mlp_hps = mlp_tuner.get_best_hyperparameters(1)[0]
        best_mlp_model = mlp_tuner.hypermodel.build(best_mlp_hps)
        # Print best hyperparameters
        print("\nBest MLP Hyperparameters:")
        for param, value in best_mlp_hps.values.items():
            print(f"{param}: {value}")
        mlp_end = time.time()
        print(f"MLP Tuning Time: {mlp_end - mlp_start:.2f} seconds")
        return best_mlp_model

    except Exception as e:
        print(f"Error - hyperparameter tuning: {str(e)}")


# --------------------------------
# Train and evaluate the models
# --------------------------------

def train_and_evaluate(mlp_model, X_train_, y_train_, X_val, y_val):
    # Train MLP
    mlp_history = mlp_model.fit(
        X_train_, y_train_,
        validation_data=(X_val, y_val),
        epochs=20,
        batch_size= 32,
        callbacks=[early_stopping],
        verbose=1
    )
    
    # Visualize training history
    plt.figure(figsize=(12, 5))

    # Plot accuracy
    plt.subplot(1, 2, 1)
    plt.plot(mlp_history.history['accuracy'], label='MLP Train')
    plt.plot(mlp_history.history['val_accuracy'], label='MLP Validation')
    plt.title('Model Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    # Plot loss
    plt.subplot(1, 2, 2)
    plt.plot(mlp_history.history['loss'], label='MLP Train')
    plt.plot(mlp_history.history['val_loss'], label='MLP Validation')
    plt.title('Model Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.tight_layout()
    plt.show()

    # Evaluate on validation set
    mlp_val_loss, mlp_val_acc = mlp_model.evaluate(X_val, y_val, verbose=0)

    print(f"MLP Validation Accuracy: {mlp_val_acc:.4f}")

    return mlp_model
```

## Convolutional neural network

```python
# Early stopping callback to prevent overfitting during tuning
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=3,
    restore_best_weights=True
)


# --------------------------------
# CNN Hyperparameter Tuning Overview
# - conv2_filters: Number of filters in 2nd Conv2D layer (64 or 128)
# - use_third_conv: Whether to add a 3rd Conv2D block
# - conv3_filters: Number of filters in optional 3rd Conv2D layer (128 or 256)
# - learning_rate: Adam optimizer learning rate (1e-3 or 1e-4)
#
# Fixed Parameters:
# - Kernel size: 3x3
# - First Conv2D block: 32 filters
# - Dense layers: 1 layer with 64 units + dropout(0.3)
# - Activation function: ReLU
# --------------------------------

def build_cnn_model(hp):
    model = Sequential()

    # Fixed kernel size for simplicity
    kernel_size = (3, 3)

    # First Conv Block (fixed parameters)
    model.add(Conv2D(
        filters=32, 
        kernel_size=kernel_size,
        activation='relu',
        padding='same',
        input_shape= (28, 28, 4)
    ))
    model.add(MaxPooling2D(pool_size=(2, 2)))

    # Second Conv Block (tuned filter size)
    model.add(Conv2D(
        filters=hp.Choice('conv2_filters', values=[64, 128]),
        kernel_size=kernel_size,
        activation='relu',
        padding='same'
    ))
    model.add(MaxPooling2D(pool_size=(2, 2)))

    # Optional third Conv Block
    if hp.Boolean('use_third_conv'):
        model.add(Conv2D(
            filters=hp.Choice('conv3_filters', values=[128, 256]),
            kernel_size=kernel_size,
            activation='relu',
            padding='same'
        ))
        model.add(MaxPooling2D(pool_size=(2, 2)))

    # Dense layers (fixed to 1 layer of 64 units)
    model.add(Flatten())
    model.add(Dense(64, activation='relu'))
    model.add(Dropout(0.3)) 

    # Output
    model.add(Dense(num_classes, activation='softmax'))

    # Tune only learning rate and fix optimizer
    learning_rate = hp.Choice('learning_rate', [1e-3, 1e-4])
    opt = tf.keras.optimizers.Adam(learning_rate=learning_rate)

    model.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])

    return model


# --------------------------------
# Run the tuners
# --------------------------------
def cnn_hyperparameter_tuning(cnn_tuner, X_train_, y_train_, X_val, y_val):
    # For debugging: print input shapes
    print(f"X_train shape: {X_train_.shape}")
    print(f"y_train shape: {y_train_.shape}")

    try:

        print("\nNow CNN hyperparameter tuning...")
        cnn_start = time.time()
        cnn_tuner.search(
            X_train_, y_train_,
            validation_data=(X_val, y_val),
            epochs=10,
            batch_size= 32,
            callbacks=[early_stopping],
            verbose=1
        )

        best_cnn_hps = cnn_tuner.get_best_hyperparameters(1)[0]
        best_cnn_model = cnn_tuner.hypermodel.build(best_cnn_hps)

        print("\nBest CNN Hyperparameters:")
        for param, value in best_cnn_hps.values.items():
            print(f"{param}: {value}")

        return best_cnn_model
        
        
    except Exception as e:
        print(f"Error - hyperparameter tuning: {str(e)}")


# --------------------------------
# Train and evaluate the models
# --------------------------------

def cnn_train_and_evaluate(cnn_model, X_train_, y_train_, X_val, y_val):

    # Train CNN
    cnn_history = cnn_model.fit(
        X_train_, y_train_,
        validation_data=(X_val, y_val),
        epochs=20,
        batch_size=32,
        callbacks=[early_stopping],
        verbose=1
    )

    # Visualize training history
    plt.figure(figsize=(12, 5))

    # Plot accuracy
    plt.subplot(1, 2, 1)
    plt.plot(cnn_history.history['accuracy'], label='CNN Train')
    plt.plot(cnn_history.history['val_accuracy'], label='CNN Validation')
    plt.title('Model Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    # Plot loss
    plt.subplot(1, 2, 2)
    plt.plot(cnn_history.history['loss'], label='CNN Train')
    plt.plot(cnn_history.history['val_loss'], label='CNN Validation')
    plt.title('Model Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.tight_layout()
    plt.show()

    # Evaluate on validation set
    cnn_val_loss, cnn_val_acc = cnn_model.evaluate(X_val, y_val, verbose=0)

    print(f"CNN Validation Accuracy: {cnn_val_acc:.4f}")

    return cnn_model
```

### 4.1 Data Processing

```python
# Run this cell first, it contain all the preprocessing/image augmentation function
def convert_to_lab(X_train, X_test):
    X_train_lab = np.zeros_like(X_train, dtype=np.float32)
    X_test_lab = np.zeros_like(X_test, dtype=np.float32)
    
    for i in range(X_train.shape[0]):
        X_train_lab[i] = cv2.cvtColor(X_train[i].astype(np.uint8), cv2.COLOR_RGB2Lab) / 255.0
       
    for i in range(X_test.shape[0]):
        X_test_lab[i] = cv2.cvtColor(X_test[i].astype(np.uint8), cv2.COLOR_RGB2Lab) / 255.0
    
    return X_train_lab, X_test_lab

def channel_wise_normalization(X_train, X_test):
    X_train_norm = X_train.astype('float32')/255.0
    X_test_norm = X_test.astype('float32')/255.0
    
    for channel in range(3):
        channel_mean = np.mean(X_train_norm[:, :, :, channel])
        channel_std = np.std(X_train_norm[:, :, :, channel])
        X_train_norm[:, :, :, channel] = (X_train_norm[:, :, :, channel] - channel_mean) / (channel_std + 1e-7)
        X_test_norm[:, :, :, channel] = (X_test_norm[:, :, :, channel] - channel_mean) / (channel_std + 1e-7)
    
    return X_train_norm, X_test_norm

def convert_to_hsv(X_train, X_test):
    X_train_hsv = np.zeros_like(X_train, dtype=np.float32)
    X_test_hsv = np.zeros_like(X_test, dtype=np.float32)
    
    for i in range(X_train.shape[0]):
        X_train_hsv[i] = cv2.cvtColor(X_train[i].astype(np.uint8), cv2.COLOR_RGB2HSV) / 255.0
    
    for i in range(X_test.shape[0]):
        X_test_hsv[i] = cv2.cvtColor(X_test[i].astype(np.uint8), cv2.COLOR_RGB2HSV) / 255.0
    
    return X_train_hsv, X_test_hsv

def extract_edge_features(X_train_rgb_norm, X_test_rgb_norm):
    def compute_edge(img):
        gray = cv2.cvtColor((img * 255).astype(np.uint8), cv2.COLOR_RGB2GRAY)
        sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
        sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)
        edge = np.sqrt(sobelx**2 + sobely**2)
        edge = edge / (edge.max() + 1e-7)
        return edge.astype(np.float32)

    N_train, H, W, _ = X_train_rgb_norm.shape
    N_test = X_test_rgb_norm.shape[0]

    X_train_combined = np.zeros((N_train, H, W, 4), dtype=np.float32)
    X_test_combined = np.zeros((N_test, H, W, 4), dtype=np.float32)

    X_train_combined[:, :, :, :3] = X_train_rgb_norm
    X_test_combined[:, :, :, :3] = X_test_rgb_norm

    # Add edge channel
    for i in range(N_train):
        X_train_combined[i, :, :, 3] = compute_edge(X_train_rgb_norm[i])
    for i in range(N_test):
        X_test_combined[i, :, :, 3] = compute_edge(X_test_rgb_norm[i])

    return X_train_combined, X_test_combined
```

```python
# for MLP 
def MLP_data_building(X_train, X_test, y_train, y_test): 
    # Step 1: Convert RGB to HSV
    X_train_lab, X_test_lab = convert_to_lab(X_train, X_test)
    # Step 2: Normalize all channels
    X_train_mlp, X_test_mlp = channel_wise_normalization(X_train_lab, X_test_lab)
    
    # one hot encoding for label (y)
    y_train_cat = to_categorical(y_train, num_classes)
    y_test_cat = to_categorical(y_test, num_classes)
    
    return X_train_mlp, X_test_mlp, y_train_cat, y_test_cat

# for CNN and gradient boost 
def CNN_data_building(X_train, X_test, y_train, y_test): 
    # Step 1: Convert RGB to HSV
    X_train_hsv, X_test_hsv = convert_to_hsv(X_train, X_test)
    # Step 2: Extract edge from HSV or RGB
    X_train_combined, X_test_combined = extract_edge_features(X_train_hsv * 255.0, X_test_hsv * 255.0)
    # Step 3: Normalize all 4 channels (HSV + edge)
    X_train_cnn, X_test_cnn = channel_wise_normalization(X_train_combined, X_test_combined)
    
    
    # one hot encoding for label (y)
    y_train_cat = to_categorical(y_train, num_classes)
    y_test_cat = to_categorical(y_test, num_classes)
    
    return X_train_cnn, X_test_cnn, y_train_cat, y_test_cat
```

```python
def best_mlp_training(X_train, X_test, y_train, y_test):

    # Split for internal validation
    X_train_, X_val, y_train_, y_val = train_test_split(
        X_train, y_train, test_size=0.2, random_state=42
    )

    # MLP model with best hyperparameters
    mlp_model = Sequential([
        Flatten(input_shape=X_train_.shape[1:]),
        Dense(256, activation='relu'),
        BatchNormalization(),
        Dropout(0.3),
        Dense(512, activation='relu'),
        BatchNormalization(),
        Dropout(0.3),
        Dense(256, activation='relu'),
        BatchNormalization(),
        Dropout(0.4),
        Dense(num_classes, activation='softmax')
    ])

    # Compile model
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
    mlp_model.compile(
        optimizer=optimizer,
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    # Train model
    mlp_model.fit(
        X_train_, y_train_,
        validation_data=(X_val, y_val),
        epochs=20,
        batch_size=32,
        verbose=1
    )

    # Evaluate on test set
    mlp_eval = mlp_model.evaluate(X_test, y_test, verbose=0)
    print(f"MLP Accuracy: {mlp_eval[1]:.4f}")

    return mlp_model
```

### Best CNN

```python
def cnn_model_training(X_train, X_test, y_train, y_test):
    X_train_, X_val, y_train_, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

    # CNN with best hyperparameters
    cnn_model = Sequential([
        Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=X_train.shape[1:]),
        MaxPooling2D((2, 2)),
        Conv2D(64, (3, 3), activation='relu', padding='same'),
        MaxPooling2D((2, 2)),
        Conv2D(256, (3, 3), activation='relu', padding='same'),
        MaxPooling2D((2, 2)),
        Flatten(),
        Dense(64, activation='relu'),
        Dropout(0.3),
        Dense(num_classes, activation='softmax')
    ])

    cnn_model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss='categorical_crossentropy', metrics=['accuracy'])
    cnn_model.fit(X_train_, y_train_, validation_data=(X_val, y_val), epochs=20, batch_size=32)

    cnn_eval = cnn_model.evaluate(X_test, y_test, verbose=0)
    print(f"CNN Accuracy: {cnn_eval[1]:.4f}")
    return cnn_model
```
